```python
# Módulo/bibliotecas utilizadas
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
# Define caminhos de input e output
DATA_PATH = '../data/sealevel/'        # diretório de dados (input)
IMG_PATH  = '../img/'                  # diretório de imagens (output)
```

```python
# Leitura dos dados tabulares utilizando o módulo pandas (pd)
cols = ['time','year','month','day','hour','sealevel','flag']    # define nomes das colunas da tabela de dados
df = pd.read_csv(DATA_PATH + 'can2004.dtf',names=cols, sep='\s+')       # faz a leitura dos dados
```

```python
df
```

```python
# Cria array de datetime64 e coloca no dataframe ds
time =  [np.datetime64(
                datetime.datetime(year=df['year'][k],
                month=df['month'][k],
                day=df['day'][k],
                hour=int(df['hour'][k])))
            for k in range(len(df['year']))]  # laço for no estilo "comprehension list"

time = np.array(time)    # transforma lista em array
df['datetime'] = time    # coloca o array no dataframe df
```

```python
df
```

```python
df.plot('datetime','sealevel')
plt.xlabel(' ')
plt.ylabel('Nível do mar [m]')
```

```python
# Remove dado suspeito (apesar da flag boa) com nível do mar acima de 7 m
df[df.sealevel>7] = np.nan
```

```python
# Plota e salva série de tempo de nível do mar (2004)
df.plot('datetime','sealevel',legend=False)
plt.ylim(0.5,3.2)
plt.xlabel(' ')
plt.ylabel('Nível do mar [m]')
plt.savefig(IMG_PATH+'niveldomar_cananeia_2004.png')
```

```python
# Seleciona um único mês
df_july = df[df.month == 7]
```

```python
# Plota e salva série de tempo de nível do mar (julho de 2004)
df_july.plot('datetime','sealevel',legend=False)
plt.ylim(0.5,3.2)
plt.xlabel(' ')
plt.ylabel('Nível do mar [m]')
plt.savefig(IMG_PATH+'niveldomar_cananeia_200407.png')
```

```python
# select single month
df_day = df_july[df_july.day == 4]
```

```python
# Plota e salva série de tempo de nível do mar (4 de julho de 2004)
fig, ax = plt.subplots()
df_day.plot('datetime','sealevel',legend=False,ax=ax)
plt.ylim(1,2.8)
ax.grid(axis='both', which='both')
plt.xlabel(' ')
plt.ylabel('Nível do mar [m]')
plt.savefig(IMG_PATH+'niveldomar_cananeia_20040707.png')
```
